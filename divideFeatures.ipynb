## Load datasets with webpage's texts

```python
phishing_texts = pd.read_csv('4_divided_by_languages/phishing/phishing_english.csv')
malware_texts = pd.read_csv('4_divided_by_languages/malware/malware_english.csv')
benignWithMalware_texts = pd.read_csv('4_divided_by_languages/benign/benign_english.csv')
benignWithPhishing_texts = pd.read_csv('4_divided_by_languages/benign/benign_english.csv')
```

## Change location of some features

```python
column1 = malware_texts.pop("word_count")
column2 = malware_texts.pop("domain_length")
malware_texts.insert(5, "word_count", column1)
malware_texts.insert(6, "domain_length", column2)

column1 = phishing_texts.pop("word_count")
column2 = phishing_texts.pop("domain_length")
phishing_texts.insert(5, "word_count", column1)
phishing_texts.insert(6, "domain_length", column2)

column1 = benignWithMalware_texts.pop("word_count")
column2 = benignWithMalware_texts.pop("domain_length")
benignWithMalware_texts.insert(5, "word_count", column1)
benignWithMalware_texts.insert(6, "domain_length", column2)

column1 = benignWithPhishing_texts.pop("word_count")
column2 = benignWithPhishing_texts.pop("domain_length")
benignWithPhishing_texts.insert(5, "word_count", column1)
benignWithPhishing_texts.insert(6, "domain_length", column2)


benignWithPhishing_texts
```

```python
import pandas as pd
from collections import Counter
import string

def count_phishing_words(text):
    """
    Counts occurrences of predefined phishing and benign words in the input text.
    Returns a dictionary with word frequency features (e.g., 'tfidf_phishing_<word>').
    """
    words = text.lower().split()
    words = [word.strip(string.punctuation) for word in words]
    counter = Counter(words)

    return {f"tfidf_phishing_{word}": counter[word] for word in combined_phish_benign_words}

def count_malware_words(text):
    """
    Counts occurrences of predefined malware and benign words in the input text.
    Returns a dictionary with word frequency features (e.g., 'tfidf_malware_<word>').
    """
    words = text.lower().split()
    words = [word.strip(string.punctuation) for word in words]
    counter = Counter(words)

    return {f"tfidf_malware_{word}": counter[word] for word in combined_malw_benign_words}


benignWithPhishing_texts['word_counts'] = benignWithPhishing_texts['text'].apply(count_phishing_words)
word_freq_df = benignWithPhishing_texts['word_counts'].apply(pd.Series)
benignWithPhishing_texts = pd.concat([benignWithPhishing_texts, word_freq_df], axis=1)

benignWithMalware_texts['word_counts'] = benignWithMalware_texts['text'].apply(count_malware_words)
word_freq_df = benignWithMalware_texts['word_counts'].apply(pd.Series)
benignWithMalware_texts = pd.concat([benignWithMalware_texts, word_freq_df], axis=1)

phishing_texts['word_counts'] = phishing_texts['text'].apply(count_phishing_words)
word_freq_df = phishing_texts['word_counts'].apply(pd.Series)
phishing_texts = pd.concat([phishing_texts, word_freq_df], axis=1)

malware_texts['word_counts'] = malware_texts['text'].apply(count_malware_words)
word_freq_df = malware_texts['word_counts'].apply(pd.Series)
malware_texts = pd.concat([malware_texts, word_freq_df], axis=1)

benignWithPhishing_texts.drop(columns=['text', 'word_counts'], inplace=True)
benignWithMalware_texts.drop(columns=['text', 'word_counts'], inplace=True)
malware_texts.drop(columns=['text', 'word_counts'], inplace=True)
phishing_texts.drop(columns=['text', 'word_counts'], inplace=True)
```

```python
benignWithPhishing_texts = benignWithPhishing_texts.loc[:, ~benignWithPhishing_texts.columns.duplicated()]
benignWithMalware_texts = benignWithMalware_texts.loc[:, ~benignWithMalware_texts.columns.duplicated()]
phishing_texts = phishing_texts.loc[:, ~phishing_texts.columns.duplicated()]
malware_texts = malware_texts.loc[:, ~malware_texts.columns.duplicated()]
```

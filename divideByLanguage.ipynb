## Filtering featues

```python
# for benign dataset i expect at least 50 words and 10 tags
# for phishing and malware at least 5 words and 1 phishing
```

```python
filtered_csnt["word_count"] = filtered_csnt["text"].apply(lambda x: len(str(x).split(" ")))
filtered_csnt["domain_length"] = filtered_csnt["domain_name"].apply(lambda x: len(str(x)))
filtered_csnt = filtered_csnt[(filtered_csnt["word_count"] >= 50) & (filtered_csnt["all_tags"] >= 10)]
    
filtered_umb["word_count"] = filtered_umb["text"].apply(lambda x: len(str(x).split(" ")))
filtered_umb["domain_length"] = filtered_umb["domain_name"].apply(lambda x: len(str(x)))
filtered_umb = filtered_umb[(filtered_umb["word_count"] >= 50) & (filtered_umb["all_tags"] >= 10)]

filtered_phish["word_count"] = filtered_phish["text"].apply(lambda x: len(str(x).split(" ")))
filtered_phish["domain_length"] = filtered_phish["domain_name"].apply(lambda x: len(str(x)))
filtered_phish = filtered_phish[(filtered_phish["word_count"] >= 5) & (filtered_phish["malicious_tags"] >= 1)]

filtered_malw["word_count"] = filtered_malw["text"].apply(lambda x: len(str(x).split(" ")))
filtered_malw["domain_length"] = filtered_malw["domain_name"].apply(lambda x: len(str(x)))
filtered_malw = filtered_malw[(filtered_malw["word_count"] >= 5) & (filtered_malw["malicious_tags"] >= 1)]
```

## Saving datasets depending on language and category

```python
import os
import pandas as pd

BENIGN_DIR = "4_divided_by_languages/benign"
PHISHING_DIR = "4_divided_by_languages/phishing"
MALWARE_DIR = "4_divided_by_languages/malware"
UNKNOWN_DIR = "4_divided_by_languages/unknown"

benign_dfs = [
    filtered_csnt,
    filtered_umb
]

phishing_dfs = [
    filtered_phish
]

malware_dfs = [
    filtered_malw
]

def save_to_csv_by_language(dfs, base_type):
    for df in dfs:
        for language in df["language"].unique():
            lang_map = {
                "en": "english",
                "es": "spanish",
                "ru": "russian",
                "fr": "french",
                "de": "german",
                "cs": "czech"
            }
            filename = f"{base_type}_{lang_map.get(language, 'unknown')}.csv"
            
            # Defining the save path
            if base_type == "benign":
                dir_path = BENIGN_DIR
            elif base_type == "phishing":
                dir_path = PHISHING_DIR
            elif base_type == "malware":
                dir_path = MALWARE_DIR
            else:
                dir_path = UNKNOWN_DIR

            filepath = os.path.join(dir_path, filename)

            # Create a folder if it doesn't exist
            os.makedirs(dir_path, exist_ok=True)

            # Select the required rows
            language_df = df[df["language"] == language]

            # Check if the file exists
            if os.path.isfile(filepath):
                # Read the existing file to avoid duplicates
                existing_df = pd.read_csv(filepath)

                # Find new rows by excluding existing ones based on 'domain_name'
                new_data = language_df.loc[~language_df['domain_name'].isin(existing_df['domain_name'])]

                if not new_data.empty:
                    # Append new data to the existing file
                    new_data.to_csv(filepath, mode='a', index=False, header=False)
            else:
                # If file does not exist, save all the data
                language_df.to_csv(filepath, mode='w', index=False, header=True)

# Start saving for all data types
save_to_csv_by_language(benign_dfs, "benign")
save_to_csv_by_language(phishing_dfs, "phishing")
save_to_csv_by_language(malware_dfs, "malware")


```
